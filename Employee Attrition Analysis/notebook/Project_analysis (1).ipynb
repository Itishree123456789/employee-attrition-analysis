# Employee Attrition Analysis Using Python

## Project Overview

Employee attrition is a key workforce metric that measures the number of employees leaving an organization. High attrition can lead to increased recruitment costs, loss of experienced talent, and reduced productivity. This project analyzes employee data to identify patterns related to attrition and understand how factors such as age, department, income, and job satisfaction influence employee turnover.

## Objective

The objective of this project is to analyze employee data and identify factors associated with employee attrition. The analysis aims to uncover trends in workforce behavior and provide insights that can help organizations improve employee retention and workforce planning.

## Dataset 

Dataset: IBM HR Analytics Employee Attrition Dataset

Columns used:

- Age
- Department
- MonthlyIncome
- JobSatisfaction
- Attrition

Dataset Link: https://www.kaggle.com/datasets/pavansubhasht/ibm-hr-analytics-attrition-dataset?utm_source=chatgpt.com

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition-selected-columns.csv")
df.head()
```

## Dataset Overview

```python
print("Dataset Shape:", df.shape)

df.info()

df.describe()
```

## Check missing values

```python
df.isnull().sum()
```

### Insight
There is no missing value present  

## Attrition Distribution

```python
attrition_count = df['Attrition'].value_counts()

print(attrition_count)

attrition_count.plot(kind='bar')

plt.title("Employee Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Count")

plt.show()
```

### Insight
Most employeess stayed with the Company,While only a small propotion left.This suggests a relatively low attrition rate and highlights the need to analyze the characteristics of employees who resigned.

## Attrition by Department

```python
department_attrition = pd.crosstab(
    df['Department'],
    df['Attrition']
)

department_attrition.plot(kind='bar')

plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Count")

plt.show()
```

### Insight

Employee attrition is most prominent in the Research & Development and Sales departments. While these departments also have larger employee populations, they contribute the majority of employee turnover, indicating that retention efforts may need to be focused on these areas.

## Employee Age Distribution

```python
plt.hist(df['Age'], bins=10)

plt.title("Employee Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

### Insight

The workforce is primarily composed of mid-career professionals, with the highest concentration of employees in the 30–40 age group. This suggests that the organization relies heavily on experienced employees who are in the middle stage of their careers

## Average Salary by Department

```python
avg_salary = df.groupby(
    'Department'
)['MonthlyIncome'].mean()

print(avg_salary)

avg_salary.plot(kind='bar')

plt.title("Average Salary by Department")
plt.ylabel("Average Monthly Income")

plt.show()
```

### Insight

Average salary levels vary across departments, with employees in Sales earning the highest monthly income. Differences in compensation may influence employee satisfaction and retention, making salary an important factor to consider when analyzing attrition trends.

## Job Satisfaction vs Attrition

```python
job_sat = pd.crosstab(
    df['JobSatisfaction'],
    df['Attrition']
)

job_sat.plot(kind='bar')

plt.title("Job Satisfaction vs Attrition")
plt.xlabel("Job Satisfaction")
plt.ylabel("Count")

plt.show()
```

### Insight

Job satisfaction appears to have a noticeable relationship with employee retention. Employees reporting lower satisfaction levels are more likely to leave the organization, while employees with higher satisfaction levels tend to remain with the company.

## Key Findings
- Out of 1,470 employees, 237 employees left while 1,233 employees remained, indicating a relatively low overall attrition rate.
- Research & Development and Sales departments account for the majority of employee turnover.
- The workforce is largely concentrated in the 30–40 age group, representing the organization's core employee base.
- Salary levels differ across departments, with Sales employees earning the highest average monthly income.
- Higher job satisfaction is associated with lower attrition, suggesting employee satisfaction plays an important role in retention.

## Conclusion

This project analyzed employee attrition patterns using workforce data from different departments. The analysis revealed that employee turnover is influenced by factors such as department, salary, and job satisfaction. Research & Development and Sales departments experienced higher attrition compared to Human Resources, while employees with lower job satisfaction were more likely to leave the organization.
